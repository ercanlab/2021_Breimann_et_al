# Violin plots 

```python
# select the data 

Sample1_plot = Sample1['length'] #wt
Sample2_plot = Sample2['length'] # DPY-21 KO
Sample3_plot = Sample3['length'] # DPY-21 jmjc
Sample4_plot = Sample4['length'] # set-4 KO
Sample5_plot = Sample5['length'] # sir-2.1 KO


Sample1_plot.reset_index(drop=True, inplace=True)
Sample2_plot.reset_index(drop=True, inplace=True)
Sample3_plot.reset_index(drop=True, inplace=True)
Sample4_plot.reset_index(drop=True, inplace=True)
Sample5_plot.reset_index(drop=True, inplace=True)

df1 = pd.DataFrame(Sample1_plot)
df2 = pd.DataFrame(Sample2_plot)
df3 = pd.DataFrame(Sample3_plot)
df4 = pd.DataFrame(Sample4_plot)
df5 = pd.DataFrame(Sample5_plot)

df1['strain']= 'wild-type'
df2['strain']= 'dpy-21 (e428)'
df3['strain']= 'dpy-21 (y607) JmjC'
df4['strain']= 'set-4 (n4600)'
df5['strain']= 'sir-2.1 (ok434)'



# Concatenate into one dataframe
df = pd.concat([df1 ,
                df2,
                df3,
               df4,
                df5],
             
                axis=0,  ignore_index=True)

df.head()
```

```python
# set colours for plotting
color_dict = {'wild-type': "#8AB17D", 
              'dpy-21 (e428)': "#970ED6", 
              'dpy-21 (y607) JmjC': "#F81CC0", 
              'set-4 (n4600)': "#0094C6", 
              'sir-2.1 (ok434)': "#929292", 
             }
```

```python
sns.reset_defaults()
fig, ax = plt.subplots(figsize=(6, 5))


ax = sns.violinplot(x="strain", y="length", data=df, 
                    palette = color_dict, 
                    linewidth= 2, 
                    width = 1.1, 
                    color = 'w')


#removing top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#labels for plot
plt.title('Length of adult worms', fontsize=15)
plt.ylabel('worm length', fontsize=15) 

#plt.ylim(0.8, 1)
#plt.xlim(-0.1,1.9)
plt.xticks(fontsize=15, rotation=40)
plt.yticks(fontsize=15)
 

plt.tight_layout()
plt.savefig((outpath + 'worm_length_violin.pdf'), dpi=400)

plt.show()
```
